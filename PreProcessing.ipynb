```python
import pandas as pd
```

```python
pddf = pd.read_csv("batting_card.csv")
```

```python
pddf.info()
```

```python
pdDF = pddf.copy()
pdDF
```

```python
pdDF.equals(pddf)
```

```python
pdDF = pdDF.drop_duplicates(keep = False)
pdDF
```

```python
def remove_whitespace(s):
    return s.strip()
pdDF = pdDF.applymap(lambda x: remove_whitespace(x) if isinstance(x, str) else x)
pdDF
```

```python
pdDF.drop(['shortText', 'match_id', 'match_name', 'venue', 'city', 'country', 'home_team', 'away_team',
           'commentary', 'link', 'season', 'captain', 'name', 'fullName', 'innings_id', 
           'current_innings'], axis=1, inplace=True)
pdDF
```

```python
pdDF= pdDF.dropna()
```

```python
pdDF['isNotOut'] = pdDF['isNotOut'].replace({True: 1, False: 0})
```

```python
split_data = pdDF['runningScore'].str.split(',', expand=True, n=2)
split_data.columns = ['Wickets', 'Runnings']
pdDF = pd.concat([pdDF, split_data], axis=1).drop('runningScore', axis=1)
```

```python
pdDF['Runnings'] = pdDF['Runnings'].str.replace('runs', '')
pdDF['Runnings'] = pdDF['Runnings'].str.replace('}', '')
pdDF['Runnings'] = pdDF['Runnings'].str.replace(':', '')
pdDF['Runnings'] = pdDF['Runnings'].str.replace("'", '')

pdDF['Wickets'] = pdDF['Wickets'].str.replace('{', '')
pdDF['Wickets'] = pdDF['Wickets'].str.replace('wickets', '')
pdDF['Wickets'] = pdDF['Wickets'].str.replace(':', '')
pdDF['Wickets'] = pdDF['Wickets'].str.replace("'", '')
pdDF
```

```python
pdDF['Wickets'] = pd.to_numeric(pdDF['Wickets'], errors='coerce').fillna(0).astype(int)
pdDF['Runnings'] = pd.to_numeric(pdDF['Runnings'], errors='coerce').fillna(0).astype(int)
pdDF['strikeRate'] = pd.to_numeric(pdDF['strikeRate'], errors='coerce').fillna(0).astype(int)
```
